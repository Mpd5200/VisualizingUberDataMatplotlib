```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)

city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data

city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

city_df = city_df.drop_duplicates('city', keep = 'first')

# Combine the data into a single dataset
merge_table = pd.merge(city_df,ride_df, on = 'city', how = 'inner' )
```

## Bubble Plot of Ride Sharing Data

```python
#group the data by city 

by_city = merge_table.groupby('city')

avg_fare = by_city['fare'].mean()

total_rides = by_city['ride_id'].count()

total_drivers = by_city['driver_count'].mean()

city_type = by_city['type'].unique()

#create data frame 

city_ride_info = pd.DataFrame ({ 'Average Fare':avg_fare,
                                 'Total Rides': total_rides,
                                 'Total Drivers': total_drivers,
                                 'City Type':city_type
                               })

city_ride_info


rural_city = city_ride_info[city_ride_info['City Type'] == 'Rural']
suburban_city = city_ride_info[city_ride_info['City Type'] == 'Suburban']
urban_city = city_ride_info[city_ride_info['City Type'] == 'Urban']


# Build the scatter plots for each city types

plt.scatter(rural_city['Total Rides'], rural_city['Average Fare'], s = rural_city['Total Drivers']*10,
edgecolor = 'black', label = 'Rural', color = 'gold', alpha = .85)

plt.scatter(suburban_city['Total Rides'], suburban_city['Average Fare'], s = suburban_city['Total Drivers']*10,
edgecolor = 'black', label = 'Suburban', color = 'lightskyblue', alpha = .85)

plt.scatter(urban_city['Total Rides'], urban_city['Average Fare'], s = urban_city['Total Drivers']*10,
edgecolor = 'black', label = 'Urban', color = 'lightcoral', alpha = .85)


# Incorporate the other graph properties

plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')
plt.grid()
plt.suptitle('Pyber Ride Sharing Data (2016)', fontsize=12, fontweight='bold')

# Create a legend

legend = plt.legend(loc="best")

#resize legend

legend.legendHandles[0]._sizes = [75]
legend.legendHandles[1]._sizes = [75]
legend.legendHandles[2]._sizes = [75]

# Incorporate a text label regarding circle size

plt.text(45,35,"Note:Circle size correlates with driver count per city")

# Save Figure

plt.savefig("../Images/Pyber_city_scatter.png")

plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents

fare_totals = merge_table.groupby('type')

total_fare = fare_totals['fare'].sum()

labels = total_fare.index

colors = ['gold', 'lightskyblue', 'lightcoral' ]

explode = [0.0,0.0,0.1]

# Build Pie Chart

plt.pie(total_fare, labels=labels, shadow = True, colors=colors, explode=explode, startangle = 140,autopct="%1.1f%%")

plt.title("% of Total Fares by City Type")
plt.axis('equal')

# Save Figure

plt.savefig("../Images/Pyber_city_fares.png")

# Show Figure
plt.show()
```

## Total Rides by City Type

```python
ride_totals = merge_table.groupby('type')

total_rides = ride_totals['ride_id'].count()

labels = total_rides.index

colors = ['gold', 'lightskyblue', 'lightcoral' ]

explode = [0.0,0.0,0.1]

# Build Pie Chart

plt.pie(total_rides, labels=labels, colors=colors, explode=explode, shadow = True,  startangle = 140,autopct="%1.1f%%")

plt.title("% of Total Rides by City Type")
plt.axis('equal')

# Save Figure

plt.savefig("../Images/Pyber_city_Rides.png")

# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
driver_totals = city_df.groupby('type')

total_driver = driver_totals['driver_count'].sum()

labels = total_driver.index

colors = ['gold', 'lightskyblue', 'lightcoral' ]

explode = [0.0,0.0,0.2]

# Build Pie Chart

plt.pie(total_driver, labels=labels, shadow = True, explode = explode, colors=colors,
startangle = 140,autopct="%1.1f%%")

plt.title("% of Total Drivers by City Type")
plt.axis('equal')

# Save Figure

plt.savefig("../Images/Pyber_city_drivers.png")

# Show Figure
plt.show()
```
